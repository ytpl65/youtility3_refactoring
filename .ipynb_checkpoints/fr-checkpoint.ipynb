## Face Similarity

```python
models = ['VGG-Face', 'Facenet', 'OpenFace', 'DeepFace', 'DeepID', 'Dlib', 'ArcFace' , 'Ensemble']
res = DeepFace.verify(helmet, tilted, enforce_detection=False)
res
```
